```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://www.chucknorrisfacts.fr/facts/top/1')
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
jokes_list = soup.find_all('div', {'class': 'card'})
len(jokes_list)
```

```python
jokes_list[7].find('p').text
```

```python
jokes_list[7].find('span').text.replace("(", "").replace("/10)", "")
```

```python
jokes_dic = {}
boucle = 0

for joke in jokes_list :
  jokes_dic.update({joke.find('p').text : joke.find('span').text.replace("(", "").replace("/10)", "")})
  boucle +=1
  print(boucle)
```

```python
jokes_dic = {}
boucle = 0

for joke in jokes_list :
  try :
    jokes_dic.update({joke.find('p').text : float(joke.find('span').text.replace("(", "").replace("/10)", ""))})

  except :
    pass

jokes_dic
```

```python
joke_df = pd.DataFrame({'joke' : jokes_dic.keys(),
                        'ratings' : jokes_dic.values()})
joke_df
```
