Imagine que tu as ce tableau NumPy mesh_data (1000 points, 4 colonnes : x, y, z, epaisseur).

```python
import numpy as np

# Création du maillage fictif
np.random.seed(42)
mesh_data = np.random.rand(1000, 4)
mesh_data[:, 3] = np.random.uniform(0.5, 3.0, 1000) # Épaisseur en mm
```

1. Isoler uniquement la colonne des épaisseurs dans une variable epaisseurs.

2. Trouver la valeur de l'épaisseur la plus faible (le point le plus critique).

3. (Optionnel) Trouver les coordonnées (x, y, z) de ce point précis.

```python
mesh_data
```

```python
epaisseur = mesh_data[:, 3]

print(f"L'epaisseur la plus faible est {epaisseur.min()}")
```

```python
index_min = np.argmin(epaisseur)
coords_min = mesh_data[index_min, :]
coords_min = coords_min[:3]
print(coords_min)
```

```python
# Maintenant que tu as coords_min (qui est un petit tableau avec [x, y, z]),
# comment calculerais-tu sa distance par rapport à l'origine (0, 0, 0) ?
```

```python
distance = np.linalg.norm(coords_min)

print(distance)
```

Exercice 3 : la Normalisation (Min-Max Scaling).

```python
temps = np.array([180, 200, 220, 190, 210])
```

La mission : Écris la ligne de code (en une seule fois si possible) pour transformer ce tableau afin que les valeurs soient comprises entre 0 et 1.

```python
def normalisation(x):
  x_norm = (x-np.min(x))/(np.max(x) - np.min(x))
  return x_norm
```

```python
normalisation(temps)
```
