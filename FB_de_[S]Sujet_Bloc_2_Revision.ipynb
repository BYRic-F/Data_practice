```python
import pandas as pd
import re

# Chargement du dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
"""
1. Combien de passagers y avait-il au total sur le Titanic ?
"""

print(f"Nombre total de pasasgers : {df.shape[0]}")
```

```python
"""
2. Combien de passagers ont survécu ? (colonne 'Survived' = 1)
"""

nombre_survivants = df['Survived'].value_counts()

print(f"Nombre total de passagers : {nombre_survivants[1]}")
```

```python
"""
3. Quel est le taux de survie (en %) ?
"""

pourcentage_survivants = df['Survived'].value_counts(normalize = True)*100

print(f"Taux de survie : {round(pourcentage_survivants[1], 2)} %")
```

```python
"""
4. Quel est l'âge moyen des passagers ? Ignore les valeurs manquantes.
"""

age_moyen = round(df['Age'].mean(), 2)

print(f"Age moyen des passagers : {age_moyen} ans")
```

```python
"""
5. Affiche le nombre de passagers par classe ('Pclass').
"""

df['Pclass'].value_counts()
```

```python
"""
6. Affiche l'âge moyen par classe ('Pclass') et par sexe ('Sex').
"""

resultat  = round(df.groupby(["Pclass", "Sex"])["Age"].mean(),2)
resultat
```

```python
"""
7. Quels sont les noms des 5 passagers les plus âgés ?
"""

passager = df[['Name', "Age"]].sort_values("Age", ascending = False).head(5)
passager
```

```python
"""
8. Crée une nouvelle colonne 'Title' extraite du nom de chaque passager (M., Mme, Miss, etc.).
"""

df['Title'] = df["Name"].apply(lambda x : re.search(r'([a-zA-Z]+(?=\.))', x).group(0))
df.sample(5)
```

```python
"""
9. Affiche la moyenne d’âge pour chaque titre ('Title').
"""

moyenne_age = round(df.groupby("Title")["Age"].mean(),0)
moyenne_age
```

```python
"""
10. Crée une colonne 'FamilySize' qui additionne 'SibSp' + 'Parch' + 1.
Puis affiche la taille de famille moyenne des survivants et des non-survivants.
"""

df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

resultat = df.groupby("Survived")["FamilySize"].mean()

resultat
```

# JSON

```python
import json

data_json = '''
[
    {
        "id": 1,
        "name": "John Smith",
        "contact": {
            "email": "john@example.com",
            "phone": "123-456-7890"
        },
        "purchases": [
            {"item": "Book", "price": 12.99},
            {"item": "Pen", "price": 1.99}
        ]
    },
    {
        "id": 2,
        "name": "Alice Johnson",
        "contact": {
            "email": "alice@example.com",
            "phone": "987-654-3210"
        },
        "purchases": [
            {"item": "Notebook", "price": 5.99}
        ]
    }
]
'''

data = json.loads(data_json)
```

```python
"""
À partir du JSON ci-dessus :
- Crée une DataFrame avec les colonnes suivantes : id, name, email, phone, total_spent
- total_spent correspond à la somme des achats de chaque client
"""
df_spent = pd.json_normalize(
    data,
    meta=["name"],
    record_path = ["purchases"])

df_total = df_spent.groupby("name")["price"].sum().reset_index()

df_name =  pd.json_normalize(data)

df_final = pd.merge(df_name, df_total, on="name").drop(columns= "purchases").rename(columns={"contact.email": "contact", "contact.phone" : "phone"})
df_final

```

```python
# Autre solution :


data = json.loads(data_json)

clients = []

for client in data:
    total_spent = 0

    for achat in client['purchases']:
        total_spent += achat['price']

    clients.append({
        'id': client['id'],
        'name': client['name'],
        'email': client['contact']['email'],
        'phone': client['contact']['phone'],
        'total_spent': total_spent
    })

df = pd.DataFrame(clients)

print(df)
```
