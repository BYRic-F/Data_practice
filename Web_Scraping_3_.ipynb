```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
# Récupération des liens actors dans casting

url = 'https://www.allocine.fr/film/fichefilm-114782/casting/'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
# Récuperation des données dans Acteurs / Actrices
soup_actor= soup.find("section",{"class" : "section casting-actor"})
print(soup_actor)
```

```python
# Récuperaiton des 8 premiers acteurs avec un encart
links_premier = soup_actor.find_all("a", {"class" : "meta-title-link"})

#Récupération des acteurs dans la liste en dessous

links_autres = (soup_actor.find_all("a", {"class" : 'item link'}))
# Certains sont enregistrés sous la class xXx item link, mais il est inutile de les recupérer car cette classe intègre des acteurs sans aucune documentation dessus, donc leur age est absent

links = links_premier + links_autres
links
```

```python
#Création du dataframe avec le nom et le lien
df_name = pd.DataFrame({"Name": [i.text for i in links],
          "link" : [i['href'] for i in links]})
df_name
```

```python
# Erreur si 2 nationalité car l'age se trouve en detail[2]
url = "https://www.allocine.fr" + "/personne/fichepersonne_gen_cpersonne=525421.html"

html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
detail = soup.find_all("div" , {"class" : "dark-grey"})
detail
```

```python
#Creation d'une fonction de récupération de l'age
def recup_age(row) :
  age = None
  name = row['Name']
  url_base = "https://www.allocine.fr"
  url_add = row['link']
  url_final = url_base + url_add
  html = requests.get(url_final)
  soup = BeautifulSoup(html.text, 'html.parser')
  try :
    try :
      detail = soup.find_all("div" , {"class" : "dark-grey"})
      age = detail[1].find("strong").text
      print(age)
    except :
      detail = soup.find_all("div" , {"class" : "dark-grey"})
      age = detail[2].find("strong").text
      print(age)
  except :
    print(f"Pas d'age trouvé pour {name}")

    pass
  return age

```

```python
df_name["Age"] = df_name.apply(recup_age, axis = 1)
```

```python
# J'ai vérifié, il n'y a effectivement pas d'age pour les None
df_name = df_name.dropna().astype({"Age" : int}).reset_index()
```

```python
age_moyen = df_name['Age'].mean()

print(f"L'age moyen de tous les acteurs est de {age_moyen :.2f} ans \n"
      f"Il n'est pas infaillible car nous avions des données manquantes")
```
